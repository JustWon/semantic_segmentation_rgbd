```python
for n, data_path in enumerate(data_paths[10:20]):
    
    color_path = os.path.join(data_path,'image',data_path.split('/')[-1]+'.jpg')
    depth_path = os.path.join(data_path,'depth_bfx', data_path.split('/')[-1]+'.png')
    label_path = os.path.join(data_path,'seg.mat')

    fig, (ax_a, ax_b, ax_c) = plt.subplots(ncols=3, figsize=(10, 5))

    color_img = m.imread(color_path)
    ax_a.imshow(color_img)
    depth_img = m.imread(depth_path)
    ax_b.imshow(depth_img)
    label_img = io.loadmat(label_path)['seginstances']
    ax_c.imshow(label_img)
```

# Dataset copy

```python
import shutil

for data_path in data_paths:
    from_color_path = os.path.join(data_path,'image',data_path.split('/')[-1]+'.jpg')
    from_depth_path = os.path.join(data_path,'depth_bfx', data_path.split('/')[-1]+'.png')
    from_label_path = os.path.join(data_path,'seg.mat')
    
    to_color_path = '/home/dongwonshin/Desktop/Datasets/SUNRGBD/SUNRGBD(tidy)/color/'
    to_depth_path = '/home/dongwonshin/Desktop/Datasets/SUNRGBD/SUNRGBD(tidy)/depth/'
    to_label_path = '/home/dongwonshin/Desktop/Datasets/SUNRGBD/SUNRGBD(tidy)/label/' + data_path.split('/')[-1]+'.mat'

    shutil.copy(from_color_path, to_color_path)
    shutil.copy(from_depth_path, to_depth_path)
    shutil.copy(from_label_path, to_label_path)
    
#     break;
```
